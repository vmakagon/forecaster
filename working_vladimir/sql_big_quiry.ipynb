```python
%%bigquery

SELECT  one.Date, one.Location_ID, one.Cloud9am, one.Cloud3pm, 
        two.Humidity9am, two.Humidity3pm,
        three.Pressure9am, three.Pressure3pm,
        four.RISK_MM, four.RainTomorrow, 
        five.MinTemp, five.MaxTemp, five.Temp9am, five.Temp3pm,
        six.Rainfall, six.Evaporation, six.Sunshine, six.RainToday,
        seven.WindGustDir, seven.WindGustSpeed, seven.WindDir9am, seven.WindDir3pm, seven.WindSpeed9am, seven.WindSpeed3pm,
        labels.Location

FROM `dev-rep-final-projects.au_weather.cloud_info` AS one

LEFT JOIN `dev-rep-final-projects.au_weather.humidity_info` AS two
ON one.Date = two.Date

LEFT JOIN `dev-rep-final-projects.au_weather.pressure_info` AS three
ON one.Date = three.Date

LEFT JOIN `dev-rep-final-projects.au_weather.rain_info` AS four
ON one.Date = four.Date

LEFT JOIN `dev-rep-final-projects.au_weather.temperature_info` AS five
ON one.Date = five.Date

LEFT JOIN `dev-rep-final-projects.au_weather.weather_info` AS six
ON one.Date = six.Date

LEFT JOIN `dev-rep-final-projects.au_weather.wind_info` AS seven
ON one.Date = seven.Date

LEFT JOIN `dev-rep-final-projects.au_weather.location_labels` AS labels
ON one.Location_ID = labels.Location_ID

LIMIT 5;
```
